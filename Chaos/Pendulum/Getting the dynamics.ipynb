```python
import matplotlib.pylab as plt
import numpy as np
from IPython.display import clear_output
```

```python
%%bash
g++ Pendulum.cpp
./a.out 10000 >data.txt
```

```python
a=np.genfromtxt("data.txt")
t=a[:,0]
theta=a[:,1]
omega = a[:,2]
plt.plot(t,theta)
plt.show()
plt.plot(theta,omega)
plt.show()
```

```python
for i in range(10,len(theta),50):
    plt.plot(theta[:i],omega[:i])
    plt.ylim(-3,1.5)
    plt.xlim(-3.5,3.5)
    plt.show()
    clear_output(wait=True)
```
