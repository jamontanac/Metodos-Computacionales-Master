```python
def fastfib(n,memo={}):
    if n==0 or n==1:
        return 1
    try:
        return memo[n]
    except KeyError:
        result = fastfib(n-1,memo)+ fastfib(n-2,memo)
        memo[n]=result
        return result
        
```

```python
def fib(n,memo={}):
    if n==0: 
        return 0
    elif  n==1:
            return 1
    elif n>0:
        try:
            return memo[n]
        except KeyError:
            result = fib(n-1,memo)+ fib(n-2,memo)
            memo[n]=result
            return result
    else:
        try:
            return memo[n]
        except KeyError:
            result = fib(n+2,memo)- fib(n+1,memo)
            memo[n]=result
            return result
```

```python
def fib(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    elif n>0:
        f_n_1=fib(1)
        f_n_2=fib(0)
        f_n=0
        for i in range(n-1):
            f_n = (f_n_1+f_n_2)
            f_n_2=f_n_1
            f_n_1=f_n
        return f_n
```
