# Fibonnaci Problem
Given the equation to generate the numbers we can define

```python
import matplotlib.pylab as plt
```

```python
def fib(n):
    if n==0:
        return 0
    if n==1:
        return 1
    else:
        return fib(n-1)+fib(n-2)
```

for measure the time we can import time just  as indicated

```python
import time
n=35
times=[]
for i in range(n):
    init=time.time()
    fib(i)
    difference_time=float(time.time()-init)
    times.append(difference_time)
```

Now we plot the times our code takes to run

```python
plt.figure(figsize=(10,7))
plt.plot(range(n),times)
plt.show()
```

Now we define our new function which computes the Fibonacci numbers using dynamic programming ( it means that we use the previous results, being the main reason why it is that fast)

```python
def fib_fast(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    elif n>0:
        f_n_1=fib(1)
        f_n_2=fib(0)
        f_n=0
        for i in range(n-1):
            f_n = (f_n_1+f_n_2)
            f_n_2=f_n_1
            f_n_1=f_n
        return f_n
```

```python
n=50
times=[]
for i in range(n):
    init=time.time()
    fib_fast(i)
    difference_time=float(time.time()-init)
    times.append(difference_time)
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(n),times)
plt.show()
```

So here we can see that this code is fast, it is that fast that the compute can not perceive the difference to see a bigger difference we can compute the Fibonacci numbers between 100000 and 200000 in steps of 10000 

```python
import time
n=200000
times=[]
for i in range(100000,n,10000):
    init=time.time()
    fib_fast(i)
    difference_time=float(time.time()-init)
    times.append(difference_time)
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(100000,n,10000),times)
plt.show()
```

To check our code is working we can compute the ratio of the numbers, we are going to calculate the first 1000 we avoid the first since it is zero

```python
n=1000
numbers=[]
for i in range(n):
    numbers.append(fib_fast(i))
ratio=[]
for i in range(10,len(numbers)-1):
    ratio.append(numbers[i+1]/numbers[i])
```

```python
plt.plot(ratio)
plt.ylim(1.6,1.63)
```

As we can see it is giving us the value of 1.6. it is not coincidence since that ratio has to give us the golden mean.
$$
\frac{1+\sqrt{5}}{2}\approx 1.618033988
$$
to see this let's plots the differences.

```python
import math
diferences=[]
golden_mean=(1+math.sqrt(5))*0.5
for i in range(len(ratio)):
    diferences.append(abs(ratio[i]-golden_mean)/golden_mean * 100)
plt.plot(diferences)
plt.yscale("log")
```

# Bonus
To solve this you only needed to change $n$ by $n+2$, so the recursion gives:
$$
F(n+2)=F(n+1)+F(n)\rightarrow F(n)=F(n+2)-F(n+1)
$$
therefore $F(-1)=F(1)+F(0)$, and the recursion function can be written as

```python
def fib(n):
    if n==0:
        return 0
    if n==1:
        return 1
    else:
        return fib(n+2)-fib(n+1)
```

# Euler's method for a first-order ODE
To do this you only needed to know the way to fill the list and the recursion was given by $  y_{k+1}=y_k+f(x_k,y_k)∗h$, so we will fill the list using the $f$ that was given

```python
import numpy as np
def F_d(x,y):
    return 2.0-np.exp(-4*x)-2*y #the function to fill the list
def solution(x):
    return 1+0.5*np.exp(-4.0*x)-0.5*np.exp(-2.0*x) # the exact solution to compare with

def euler(x_0,y_0,n,T=1): # Euler method which turs out to be only a way to fill the list
    y_list=[]
    x_list=[]
    y=y_0
    for i in range(n+1):
        y_list.append(y)
        x=x_0+i*(T/n)
        x_list.append(x)
        y+=(T/n)*F_d(x,y)
    return np.array(y_list),np.array(x_list)
```

Here the difficulty was not in the part of filling the list with the solutions of the Euler method, here the real challenge was to truncate the result to 6 decimals, to do this it is necessary to pass the number to string and then take the first 6 decimals, which is enough to take for this case the first 8 numbers, and then pass again the result to a float and return it

```python
def ex_euler(n):
    print(n)
    y_k,x_k=euler(0,1,n)
    z_k=solution(x_k)
    A_k_error=np.abs(y_k-z_k)/z_k
    error=np.sum(A_k_error)/(n+1)
    print("complete error",error)
    result=str(error)[:8]
    print("truncated error",result)
    return float(result)
```

# Strings
this one is kind of tricky but it is enough if you understand the rules of the game. since it always have to open and close to return True, we can always check if this is the case and then delete them and check the next. If the string at the end is empty is is because it was well defined. 

```python
def validBraces(string):
    while "()" in string or "[]" in string or "{}" in string:
        string=string.replace("()","")
        string=string.replace("[]","")
        string=string.replace("{}","")
    return string==""
```

# List of lists
this is the easier of all exercises and to do this you can make use of the fact that list have some operations defined over them

```python
def List_of_lists(list_to_work):
    sorted_list=sorted(list_to_work)
    result=[]
    for i in sorted_list:
        result.append([i]*int(i))
    return result
```

# Sorting things in other way:
this can be solve easily by using the function of numpy ``np.binary_repr`` and using the function ``count`` of strings. The solution is given below

```python
import numpy as np
def sort_by_bit(arr): 
    total={}
    a=sorted(arr)
    for i in range(len(a)):
        total[i]=str(np.binary_repr(a[i],32)).count("1")
    result=[]
    for i in sorted(total,key=total.get):
        result.append(a[i])
    return result
```

or in a shorter and a clever way but making use of a lambda function of python

```python
def sort_by_bit(a):
    return sorted(a, key=lambda x: (bin(x).count("1"), x))
```

in this problem the only difficult thing to do was to make the table that tell you the possibilities of codons for a given letter in the protein (Aminoacid).

```python
Protein_Rna={'I': ['ATT', ' ATC', '  ATA'],'L': ['CTT', '  CTC', ' CTA', ' CTG', ' TTA', ' TTG'],
             'V': ['GTT', ' GTC', ' GTA', '  GTG'],'F': ['TTT', ' TTC'],'M': ['ATG'],'C': ['TGT', '  TGC'],
             'A': ['GCT', ' GCC', ' GCA', ' GCG'],'G': ['GGT', ' GGC', ' GGA', ' GGG'],'P': ['CCT', ' CCC', ' CCA', '  CCG'],
             'T': ['ACT', ' ACC', ' ACA', '  ACG'],'S': ['TCT', ' TCC', ' TCA', ' TCG', '  AGT', ' AGC'],
             'Y': ['TAT', ' TAC'],'W': ['TGG'],'Q': ['CAA', ' CAG'],'N': ['AAT', ' AAC'],'H': ['CAT', ' CAC'],
             'E': ['GAA', ' GAG'],'D': ['GAT', ' GAC'],'K': ['AAA', ' AAG'],
             'R': ['CGT', ' CGC', ' CGA', ' CGG', '  AGA', ' AGG'],'Stop': ['TAA', ' TAG', ' TGA']}
```

Or if by any chance you have worked with pandas it has a nice way to read tables from Internet, [here](http://algoart.com/aatable.htm) you can check the table with the data you need. here for this case we are going to count the possibilities of codons for each letter

```python
import pandas as pd
df=pd.read_html("http://algoart.com/aatable.htm")[1]
df.columns = df.iloc[0]
df=df.drop(df.index[0])
df=df.drop(df.index[21])
df=df.reset_index(drop=True)
df["DNA  codons"]=df["DNA  codons"].str.split(",")
Protein_Rna_pandas={}
for i in range(len(df)):
    Protein_Rna_pandas[df["SLC"].iloc[i]]=len(df["DNA  codons"].iloc[i])
```

with this we are ready to compute the number of possibilities for a given string of Protein

```python
def From_protein_to_rna(Protein):
    result = Protein_Rna_pandas["Stop"] # We take into account the one that says that it stop
    for i in Protein:
        result*=Protein_Rna_pandas[i]
    return result
```
