# Derivación por medio de variable compleja

Una de las alternativas de derivación numerica es haciendo uso de variable compleja, se tiene que si una función $f$ es una función que cumple ciertas reglas ( una función analitica en el plano complejo), se tiene que la primera derivada puede ser calculada como un paso sobre el eje imaginario, esto es:

$$f'(x) \approx \frac{\mathfrak{Im}(f(x+ih))}{h}$$

## 1 Ejercicio
Haga uso de esta aproximación para calcular la derivada primera de la función $\sin^{5}(x)$ en $x=\pi/2$

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import cmath
```

```python
def dF_complex(x,h):
    '''Complex step aproximation'''
    #x =  np.array(x, dtype='cfloat')
    return np.imag(f(x + 1j*h))/h
def f(x):
    return np.sin(x)**5
```

```python
h=np.logspace(-6,-2)
print(dF_complex(np.pi/2.0*np.ones(len(h)),h))
```

## 2 Ejercicio
__a).__ Use la definición anterior para comparar los metodos de "Finite forward approximation", "Finite central approximation" y el metodo anterior, prepare una gráfica del error de cada metodo en función del ancho.

__b).__ Ahora haga uso de la extrapolación de Richardson sobre cada metodo y observe como cambia el error, para esto prepare una gráfica.

La función a derivar será:
$$F(x)=\frac{e^x}{\cos^3(x)+\sin^3(x)}$$
en $x=\pi/4$

```python
def F(x):
    return np.exp(x)/((np.cos(x))**3 + np.sin(x)**3)
```

```python
N = 1000

x = np.linspace(-np.pi/4,np.pi/2, N, endpoint=True) 
y =  F(x)

point_x = np.pi/4
point_y = F(point_x)
```

```python
plt.figure()
plt.plot(x,y)
plt.plot(point_x, point_y, 'o')

plt.xlim((-np.pi/4, np.pi/2))
plt.ylim((0, 6))

plt.ylabel('F(x)')
plt.xlabel('x')
plt.annotate(xy=(-0.5,4),s=r'$F(x)=\frac{e^x}{\cos(x)^3+\sin(x)^3}$',fontsize=20)

plt.plot()
plt.show()
```

```python
def dF_complex(x,h):
    '''Complex step aproximation'''
    #x =  np.array(x, dtype='cfloat')
    return np.imag(F(x + 1j*h))/h

def dF_ff(x,h):
    '''Finite forward-difference approximation'''
    return (F(x+h) - F(x))/h

def dF_cf(x,h):
    '''Finite central-difference approximation'''
    return (F(x+h) - F(x-h))/(2*h)

def dF_analytic(x):
    '''Analytic derivative (to obtain the exact value)'''
    return ((np.exp(x)*(np.cos(3*x) + np.sin(3*x)/2 + (3*np.sin(x))/2)) / 
            (np.cos(x)**3 + np.sin(x)**3)**2)
```

```python
point_x = np.pi/4
h_values = np.logspace(-15,-1,15, endpoint=True)

# Preallocate values in memory
error_complex = np.zeros(h_values.shape)
error_cf = np.zeros(h_values.shape)
error_ff = np.zeros(h_values.shape)

for i, h in enumerate(h_values):
    error_complex[i] = np.abs(dF_complex(point_x, h) - dF_analytic(point_x))/np.abs(dF_analytic(point_x))
    error_ff[i] = np.abs(dF_ff(point_x, h) - dF_analytic(point_x))/np.abs(dF_analytic(point_x))
    error_cf[i] = np.abs(dF_cf(point_x, h) - dF_analytic(point_x))/np.abs(dF_analytic(point_x))
```

```python
plt.figure()
plt.loglog(h_values, error_complex, label='Complex step')
plt.loglog(h_values, error_ff, label='Forward difference')
plt.loglog(h_values, error_cf, label='Central-difference')

plt.gca().invert_xaxis()
plt.legend()
plt.ylabel('Normalized error')
plt.xlabel('Step size (h)')

plt.show()
```

# Ejercicio 3
La formula de arriba es sólo valida para calcular la primera derivada, una generalización de esta integral está dada por la formula de la integral de Cauchy
$$f^{(n)}(x_0)=\frac{n!}{2\pi i}\oint_{\gamma} = \frac{f(z)}{(z-x_0)^{n+1}}dz,$$
esta integral puede ser calculada haciendo uso de un remplazo de variable.
$$f^{(n)}(x_0)=\frac{n!}{2\pi r^{n}}\int_{0}^{2\pi}  f(x_0+re^{i\theta})e^{-ni\theta}d\theta$$
Su tarea será implementar este metodo para calcular la derivada $30$ de las siguientes funciones $e^{-x}$ en $x=1$, $sin(x)$ y $cos(x)$ en $x=\pi/4$.

```python
def myDerivative(f, x, dTheta, degree, contourRadius):
    riemannSum = 0 
    theta = 0
    while theta < 2*np.pi:
        functionArgument = np.complex128(x + contourRadius*np.exp(1j*theta))
        secondFactor = (1/contourRadius)**degree*np.complex128(np.exp(-1j * degree * theta))
        riemannSum += f(functionArgument) * secondFactor * dTheta
        theta += dTheta

    return math.factorial(degree) * riemannSum.real / (2*np.pi)
def f(x):
    return np.sin(x)
```

$$\frac{d^{n}}{dx^{n}}\sin(x) = \sin\left(x + \frac{n\pi}{2}\right)$$

```python
h_values = np.logspace(-5,-2,15, endpoint=True)
```

```python
error=[]
for i in h_values:
    real=f(np.pi/4.0 + 30*np.pi*0.5)
    error.append(np.abs((myDerivative(f,np.pi/4.0,i,30,100)-real)/real))
```

```python
plt.plot(h_values,error)
plt.ylabel('Normalized error')
plt.xlabel(r'Step size ($d\theta$)')
plt.loglog()
plt.gca().invert_xaxis()
plt.show()
```

```python
def f(x):
    return np.exp(-x)
h_values = np.logspace(-5,-2,15, endpoint=True)
error=[]
for i in h_values:
    real=f(1)
    error.append(np.abs((myDerivative(f,1,i,30,100)-real)/real))
```

```python
plt.plot(h_values,error)
plt.ylabel('Normalized error')
plt.xlabel(r'Step size ($d\theta$)')
plt.loglog()
plt.gca().invert_xaxis()
plt.show()
```
