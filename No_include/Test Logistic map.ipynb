# Logistic map
The logistic map is a one-dimensional discrete time dynamical system that is defined by the equation
$x_{n+1}=f(x_n)=\lambda x_n (1-x_n)$

For an initial value $0\leq x_0\leq 1$ this map generates a sequence of values $x_0,x_1,\ldots ,x_n, x_{n+1}\ldots$ The growth parameter is chosen to be
$$0<\lambda\leq 4$$

```python
import numpy as np
import matplotlib.pylab as plt
def Logistic_map(x,L):
    return L*x*(1-x)
```

```python
x_total=np.linspace(0.,1.0)
plt.figure(figsize=(10,10))
plt.plot(x_total,Logistic_map(x_total,2))
plt.plot(x_total,x_total)
plt.xlim(0,1)
plt.ylim(0,1)
x=0.3
plt.plot((x,x),(0,Logistic_map(x,2)))
plt.plot((x,Logistic_map(x,2)),(Logistic_map(x,2),Logistic_map(x,2)))
```

```python
from IPython.display import clear_output
```

```python
for L in np.linspace(1,4):
    x_total=np.linspace(0.,1.0)
    x=0.001
    plt.figure(figsize=(10,10))
    plt.title(r"$\lambda=${}".format(round(L,3)))
    plt.plot(x_total,Logistic_map(x_total,L),color="navy")
    plt.plot(x_total,x_total,color="orange")
    plt.xlim(0,1)
    plt.ylim(0,1)
    for i in range(100):
        x_old=x
        x=Logistic_map(x,L)
        if i ==0:
            plt.plot((x_old,x_old),(0,x),color="firebrick")
        else:
            plt.plot((x_old,x_old),(x_old,x),color="firebrick")
        plt.plot((x_old,x),(x,x),color="firebrick")
    plt.show()
    clear_output(wait=True)
    
     
```

```python
x_total=np.linspace(0.,1.0)
x=0.001
L=3.1
for i in range(100):
    plt.figure(figsize=(10,10))
    x_old=x
    x=Logistic_map(x,L)
    if i ==0:
        plt.plot((x_old,x_old),(0,x),color="firebrick")
    else:
        plt.plot((x_old,x_old),(x_old,x),color="firebrick")
    plt.plot((x_old,x),(x,x),color="firebrick")
    
    plt.title(r"$\lambda=${}".format(round(L,3)))
    plt.plot(x_total,Logistic_map(x_total,L),color="navy")
    plt.plot(x_total,x_total,color="orange")
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.show()
    clear_output(wait=True)
```

```python
for L in np.linspace(3,4):
    x=[0.2]
    for i in range(10000):
        x.append(Logistic_map(x[-1],L))
    plt.step(x,x,alpha=1,color="rebeccapurple",linewidth=0.01)
    plt.title(r"$\lambda=${}".format(round(L,3)))
    plt.plot(x_total,Logistic_map(x_total,L),color="navy")
    plt.plot(x_total,x_total,color="orange")
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.show()
    clear_output(wait=True)
```

```python
plt.figure(figsize=(20,20))
for L in np.linspace(1,4,1000):
    x=np.linspace(0.0001,1,10000)
    for i in range(100):
        x=Logistic_map(x,L)
    plt.plot([L]*len(x),x,".",markersize=0.01,color="k")
    plt.ylim(0,1)
    plt.xlim(1.5,4)
plt.show()
```
